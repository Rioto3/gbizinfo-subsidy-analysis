# 既知の不正受給事例とgBizINFO APIデータの整合性検証

このノートブックでは**メディアで報道された不正受給事例**と**gBizINFO APIデータ**を照合し、
オープンデータの信頼性・有用性を検証します。

## 検証事例
1. カネヨ山野辺水産（宮城県塩釜市）- 復興支援補助金不正受給
2. ビック・ママ（宮城県）- 雇用調整助成金不正受給
3. ファイブエム商事（宮城県）- 雇用調整助成金不正受給

## 検証手法
- 報道された不正受給額とAPI記録の比較
- 関連補助金データの有無確認
- 不一致原因の調査

## 主要な問い
- APIデータは既知の不正を検出できるか？
- 不正受給を示す統計的特徴はあるか？
- データの限界はどこにあるか？
